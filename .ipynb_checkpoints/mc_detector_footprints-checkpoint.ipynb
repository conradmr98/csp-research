```python
%matplotlib inline
import math as m; import numpy as np; import matplotlib.pyplot as plt; import matplotlib.colors
import pandas as pd; import os; import pathlib; import string; import datetime; import h5py

directory = 'D:\\Jupiter_satellites_data\\'
```

```python
def arctan_fix(y_array, x_array):
    array_arctan = np.array([])
    for i in np.arange(len(x_array)):
        if x_array[i] >= 0:
            i_arctan = np.arctan(y_array[i]/x_array[i])
        elif x_array[i] < 0:
            i_arctan = np.pi + np.arctan(y_array[i]/x_array[i])
        array_arctan = np.append(array_arctan, i_arctan)
        
    return array_arctan
```

```python
# define PJ data filename
PJ_num = '00'
file = 'JunoFoot_PJ' + PJ_num + '_20200126_all'

# list of key names for data to extract
key_list = ['SouthFootWLonUV','SouthFootLatUV','NorthFootWLonUV','NorthFootLatUV', \
            'SouthIoFootWLonUV','SouthIoFootLatUV','NorthIoFootWLonUV','NorthIoFootLatUV', \
            'TimeUTC']

filename = directory + file + '.hdf5'
with h5py.File(filename, "r") as f:
    # print key names for all data  
    print("Keys: %s" % f.keys())

    dt_array = np.empty((0,len(f[list(f.keys())[0]][()])), dtype=datetime.datetime)
    ft_array = np.empty((0,len(f[list(f.keys())[0]][()])), dtype=float)
    
    for i in np.arange(len(key_list)):
        data_key_i = list(f.keys())[list(f.keys()).index(key_list[i])]
        data_row_i = f[data_key_i][()]  # returns as a numpy array for row i
        
        if type(f[data_key_i][()][0]) == np.float64:
            data_row_i = [f[data_key_i][()]]
            ft_array = np.concatenate((ft_array, data_row_i), axis=0)
            
        if type(f[data_key_i][()][0]) == np.bytes_:
            data_row_i = [[datetime.datetime.strptime(bytes_str.decode('utf-8'), '%Y-%m-%dT%H:%M:%S.%fZ') \
                          for bytes_str in f[data_key_i][()]]]
            dt_array = np.concatenate((dt_array, data_row_i), axis=0)
            
        if (type(f[data_key_i][()][0]) != np.bytes_) & (type(f[data_key_i][()][0]) != np.float64):
            print('DATATYPE ISSUE -> CODE BREAK')
            break
```

```python
# create time arrays
yyyy_array = np.array([]); mm_array = np.array([]); dd_array = np.array([]); hr_array = np.array([]); mn_array = np.array([])
for i in np.arange(len(dt_array[0,:])):
    yyyy_array = np.append(yyyy_array, dt_array[0,i].year); mm_array = np.append(mm_array, dt_array[0,i].month);
    dd_array = np.append(dd_array, dt_array[0,i].day); hr_array = np.append(hr_array, dt_array[0,i].hour);
    mn_array = np.append(mn_array, dt_array[0,i].minute)

print(f'PJ{PJ_num} Start Date: {mm_array[0]:.0f}/{dd_array[0]:.0f}/{yyyy_array[0]:.0f} | {hr_array[0]:.0f}:{mn_array[0]:.0f}')
print(f'PJ{PJ_num} End Date: {mm_array[-1]:.0f}/{dd_array[-1]:.0f}/{yyyy_array[-1]:.0f} | {hr_array[-1]:.0f}:{mn_array[-1]:.0f}')
```

```python
# juno north footprint coords
r_juno_n = (90-ft_array[3,:])
theta_juno_n = (ft_array[2,:])*(np.pi/180)

# io north footprint coords
r_io_n = (90-ft_array[7,:])
theta_io_n = (ft_array[6,:])*(np.pi/180)

# convert to cartesian
x_juno_n = r_juno_n * np.cos(theta_juno_n); y_juno_n = r_juno_n * np.sin(theta_juno_n)
x_io_n = r_io_n * np.cos(theta_io_n); y_io_n = r_io_n * np.sin(theta_io_n)

# take difference in cartesian
dx_footprint_n = x_juno_n - x_io_n
dy_footprint_n = y_juno_n - y_io_n

# find position of juno FT rel to io FT
theta_footprint_n = arctan_fix(dy_footprint_n, dx_footprint_n)
d_footprint_n = np.sqrt((dx_footprint_n)**2+(dy_footprint_n)**2)
```

```python
# FT crossing criteria
d_FT_max = 5 # jovigraphic degrees

ft_xing_times = np.array([])
for i in np.arange(len(d_footprint_n)):
    if d_footprint_n[i] < d_FT_max:
        t_str = f'{dt_array[0,i].year}-{dt_array[0,i].month:02}-{dt_array[0,i].day:02}/{dt_array[0,i].hour:02}:{dt_array[0,i].minute:02}'
        ft_xing_times = np.append(ft_xing_times, t_str)
        
#for i in np.arange(len(ft_xing_times))[1:-1]:
#    if dt_array[0,i-1].minute
        
print(ft_xing_times)
```

```python
# plot juno and io footprints on polar plot
tick1 = np.min(90-ft_array[3,:])
tick2 = np.min(90-ft_array[3,:]) + (np.max((90-ft_array[3,:])[ft_array[3,:]>0]) - np.min(90-ft_array[3,:]))/3
tick3 = np.min(90-ft_array[3,:]) + (np.max((90-ft_array[3,:])[ft_array[3,:]>0]) - np.min(90-ft_array[3,:]))*(2/3)
tick4 = np.max((90-ft_array[3,:])[ft_array[3,:]>0])

fig, ax = plt.subplots(figsize=(6,6),subplot_kw={'projection': 'polar'})
ax.plot(theta_juno_n, r_juno_n, color='b')
ax.plot(theta_io_n, r_io_n, color='r')
ax.set_rmax(2)
ax.set_rticks([tick1, tick2, tick3, tick4])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)
ax.set_title(f'Juno (blue) and Io (red) North Pole Footprints | PJ{PJ_num} | {mm_array[0]:.0f}/{dd_array[0]:.0f}/{yyyy_array[0]:.0f}', va='bottom')

# plot footprint distance (juno-io) on polar plot centered on io footprint
tick1 = np.min(d_footprint_n[d_footprint_n<50])
tick2 = np.min(d_footprint_n[d_footprint_n<50]) + (np.max(d_footprint_n[d_footprint_n<50]) - np.min(d_footprint_n[d_footprint_n<50]))/3
tick3 = np.min(d_footprint_n[d_footprint_n<50]) + (np.max(d_footprint_n[d_footprint_n<50]) - np.min(d_footprint_n[d_footprint_n<50]))*(2/3)
tick4 = np.max(d_footprint_n[d_footprint_n<50])
      
fig, ax = plt.subplots(figsize=(6,6),subplot_kw={'projection': 'polar'})
ax.plot(theta_footprint_n, d_footprint_n, color='purple')
ax.plot(np.linspace(0, 2*np.pi), np.array([d_FT_max]*50), ls='--', color='k')
ax.set_rmax(2)
ax.set_rticks([tick1, tick2, tick3, tick4])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)
ax.set_title(f'North Pole Footprint Distance (Io-centered) | PJ{PJ_num} | {mm_array[0]:.0f}/{dd_array[0]:.0f}/{yyyy_array[0]:.0f}', va='bottom');
```
