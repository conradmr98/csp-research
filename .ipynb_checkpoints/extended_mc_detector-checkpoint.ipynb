```python
%matplotlib inline
import math as m; import numpy as np; import matplotlib.pyplot as plt; import matplotlib.colors
import pandas as pd; import os; import pathlib; import string; import datetime; import h5py

directory = 'D:\\Jupiter_satellites_data\\'
```

```python
# moon data
sat_list = ['Io', 'Europa', 'Ganymede']
d_eq_VAR = [100, 100, 100]

RI = 0.0261 # Io radius in RJ
RE = 0.0223 # Europa radius in RJ
RG = 0.0377 # Ganymede radius in RJ

R_list = np.array([RI, RE, RG]); R_labels = np.array(['RI', 'RE', 'RG'])
years = np.arange(2016, 2023+1)
```

```python
# M-Shell data from Xiaochen code
# read in Juno M-shell data

yyyy_XC = np.array([]); mm_XC = np.array([]); dd_XC = np.array([]); hh_XC = np.array([]); mn_XC = np.array([]);
MSHELL_Juno = np.array([]); MSHELL_Io = np.array([]); MSHELL_Europa = np.array([]); MSHELL_Ganymede = np.array([]);

for year in years:
    XC_file = f'JunoMoonsMshells_Con2020_{year}.txt'
    XC_fields = ['Year', 'Month', 'Day', 'Hour', 'Minute', 'Juno_M', 'Io_M', 'Europa_M', 'Ganymede_M']
    XC_data = pd.read_csv(directory + XC_file, names = XC_fields, usecols = XC_fields, sep = '\s+',header = 0)

    yyyy_XC_i = XC_data['Year'].values; mm_XC_i = XC_data['Month'].values; dd_XC_i = XC_data['Day'].values;
    hh_XC_i = XC_data['Hour'].values; mn_XC_i = XC_data['Minute'].values; MSHELL_Juno_i = XC_data['Juno_M'].values;
    MSHELL_Io_i = XC_data['Io_M'].values; MSHELL_Europa_i = XC_data['Europa_M'].values;
    MSHELL_Ganymede_i = XC_data['Ganymede_M'].values
    
    yyyy_XC = np.append(yyyy_XC, yyyy_XC_i); mm_XC = np.append(mm_XC, mm_XC_i); dd_XC = np.append(dd_XC, dd_XC_i); 
    hh_XC = np.append(hh_XC, hh_XC_i); mn_XC = np.append(mn_XC, mn_XC_i);
    MSHELL_Juno = np.append(MSHELL_Juno, MSHELL_Juno_i); MSHELL_Io = np.append(MSHELL_Io, MSHELL_Io_i); 
    MSHELL_Europa = np.append(MSHELL_Europa, MSHELL_Europa_i); MSHELL_Ganymede = np.append(MSHELL_Ganymede, MSHELL_Ganymede_i);
```

```python
# obtain MLAT and MLT for Juno, Io, Europa, Ganymede
# read in and define MLAT and MLT data from Xiaochen data file

yyyy_pos = np.array([]); mm_pos = np.array([]); dd_pos = np.array([]); hh_pos = np.array([]); mn_pos = np.array([]);
MLAT_Juno = np.array([]); MLT_Juno = np.array([]); MLAT_Io = np.array([]); MLT_Io = np.array([]);
MLAT_Europa = np.array([]); MLT_Europa = np.array([]); MLAT_Ganymede = np.array([]); MLT_Ganymede = np.array([]);

for year in years:
    pos_file = f'jupiterMoonsMLTMLATInfo_{year}_v1.0.txt'
    pos_fields = ['Year', 'Month', 'Date', 'Hour', 'Minute', 'MLAT_Juno [degree]', 'MLT_Juno [hr]','MLAT_Io [degree]', \
                  'MLT_Io [hr]','MLAT_Europa [degree]', 'MLT_Europa [hr]', 'MLAT_Ganymede [degree]', 'MLT_Ganymede [hr]']

    pos_data = pd.read_csv(directory + pos_file, names = pos_fields, usecols = pos_fields, sep = '\s+', header = 10)
    
    # add to full list
    yyyy_pos_i = pos_data['Year'].values; mm_pos_i = pos_data['Month'].values;
    dd_pos_i = pos_data['Date'].values; hh_pos_i = pos_data['Hour'].values; mn_pos_i = pos_data['Minute'].values;
        
    MLAT_Juno_i = pos_data['MLAT_Juno [degree]'].values; MLT_Juno_i = pos_data['MLT_Juno [hr]'].values;
    MLAT_Io_i = pos_data['MLAT_Io [degree]'].values; MLT_Io_i = pos_data['MLT_Io [hr]'].values;
    MLAT_Europa_i = pos_data['MLAT_Europa [degree]'].values; MLT_Europa_i = pos_data['MLT_Europa [hr]'].values;
    MLAT_Ganymede_i = pos_data['MLAT_Ganymede [degree]'].values; MLT_Ganymede_i = pos_data['MLT_Ganymede [hr]'].values;
    
    yyyy_pos = np.append(yyyy_pos, yyyy_pos_i); mm_pos = np.append(mm_pos, mm_pos_i);
    dd_pos = np.append(dd_pos, dd_pos_i); hh_pos = np.append(hh_pos, hh_pos_i); mn_pos = np.append(mn_pos, mn_pos_i);
    MLAT_Juno = np.append(MLAT_Juno, MLAT_Juno_i); MLT_Juno = np.append(MLT_Juno, MLT_Juno_i); 
    MLAT_Io = np.append(MLAT_Io, MLAT_Io_i); MLT_Io = np.append(MLT_Io, MLT_Io_i);
    MLAT_Europa = np.append(MLAT_Europa, MLAT_Europa_i); MLT_Europa = np.append(MLT_Europa, MLT_Europa_i);
    MLAT_Ganymede = np.append(MLAT_Ganymede, MLAT_Ganymede_i); MLT_Ganymede = np.append(MLT_Ganymede, MLT_Ganymede_i); 
```

```python
for i, Sat in zip(np.arange(len(sat_list)), sat_list):
    print('Criteria :   R_sat')
    print(f'{Sat} Magnetic Conjunctions (2016-2024)\n---------------------------------------------------------------------------')
        
    x_Sat = MSHELL_Sat[i]*np.cos(MLT_Sat_trim[i]*(np.pi/12))/R_list[i]
    y_Sat = MSHELL_Sat[i]*np.sin(MLT_Sat_trim[i]*(np.pi/12))/R_list[i]
    x_Juno = MSHELL_Juno*np.cos(MLT_Juno_trim*(np.pi/12))/R_list[i]
    y_Juno = MSHELL_Juno*np.sin(MLT_Juno_trim*(np.pi/12))/R_list[i]

    d_eq = np.sqrt((x_Juno - x_Sat)**2 + (y_Juno - y_Sat)**2)
    mask = d_eq <= d_eq_VAR[i]
    for j in np.arange(len(d_eq[mask])):
        print(f'{j+1:3}) {int(yyyy_XC[mask][j])}-{int(mm_XC[mask][j]):02}-{int(dd_XC[mask][j]):02}/'\
              f'{int(hh_XC[mask][j]):02}:{int(mn_XC[mask][j]):02} | d_eq = {d_eq[mask][j]:6.2f} | '
              f'Juno_MLAT = {MLAT_Juno_trim[mask][j]:6.2f} | Juno_MLT = {MLT_Juno_trim[mask][j]:6.2f} | '
              f'Juno_MShell = {MSHELL_Juno[mask][j]:6.2f}')
    print('\n')
```

```python
print('Criteria : M_diff < 1 | MLT_diff < 1')
for i, Sat in zip(np.arange(len(sat_list)), sat_list):
    print(f'{Sat} Magnetic Conjunctions (2016-2022)\n---------------------------------------------------------------------------')
        
    x_Sat = MSHELL_Sat[i]*np.cos(MLT_Sat_trim[i]*(np.pi/12))/R_list[i]
    y_Sat = MSHELL_Sat[i]*np.sin(MLT_Sat_trim[i]*(np.pi/12))/R_list[i]
    x_Juno = MSHELL_Juno*np.cos(MLT_Juno_trim*(np.pi/12))/R_list[i]
    y_Juno = MSHELL_Juno*np.sin(MLT_Juno_trim*(np.pi/12))/R_list[i]

    d_eq = np.sqrt((x_Juno - x_Sat)**2 + (y_Juno - y_Sat)**2)
    M_diff = np.abs(MSHELL_Juno - MSHELL_Sat[i])
    
    MLT_J = MLT_Juno_trim
    MLT_S = MLT_Sat_trim[i]
    MLT_diff_gt12 = np.abs(MLT_J - MLT_S) > 12
    Juno_MLT_less = MLT_J < MLT_S
    Sat_MLT_less = MLT_S < MLT_J
    
    MLT_J[MLT_diff_gt12 & Juno_MLT_less] += 24
    MLT_S[MLT_diff_gt12 & Sat_MLT_less] += 24
    
    mask = (np.abs(MSHELL_Juno - MSHELL_Sat[i]) <= 1) & (np.abs(MLT_J - MLT_S) <= 1)

    group_end_indices = np.array([0])
    for j in np.arange(1,len(d_eq[mask])):
        if j == np.max(np.arange(1,len(d_eq[mask]))):
            group_end_indices = np.append(group_end_indices,j)
            break
        if datetime.datetime(int(yyyy_XC[mask][j+1]),int(mm_XC[mask][j+1]),int(dd_XC[mask][j+1]),int(hh_XC[mask][j+1]),\
                             int(mn_XC[mask][j+1])) - datetime.datetime(int(yyyy_XC[mask][j]),int(mm_XC[mask][j]),\
                                                                        int(dd_XC[mask][j]),int(hh_XC[mask][j]),\
                                                                        int(mn_XC[mask][j])) > datetime.timedelta(seconds=60):
            group_end_indices = np.append(group_end_indices,j)
    group_end_indices = group_end_indices[1:]
    group_start_indices = np.append(np.array([0]), group_end_indices[0:-1]+1)
    
    final_indices = np.array([0])
    for k in np.arange(len(group_start_indices)):
        k_indices = np.arange(group_start_indices[k],group_end_indices[k]+1)
        M_diff_k = M_diff[mask][k_indices]
        if len(M_diff_k) == 1:
            final_indices = np.append(final_indices, k_indices[0])
        elif len(M_diff_k) == 2:
            final_indices = np.append(final_indices, k_indices[M_diff_k == np.min(M_diff_k)])
        elif len(M_diff_k) >= 3:
            if M_diff_k[0] < M_diff_k[1]:
                final_indices = np.append(final_indices, k_indices[0])
            if M_diff_k[-1] < M_diff_k[-2]:
                final_indices = np.append(final_indices, k_indices[-1])
            for l in np.arange(1,len(M_diff_k)-1):
                if M_diff_k[l] < M_diff_k[l-1] and M_diff_k[l] < M_diff_k[l+1]:
                    final_indices = np.append(final_indices, k_indices[l])
    final_indices = final_indices[1:]
#     print(group_end_indices)
#     print(group_start_indices)
#     print(final_indices)

    count=1
    for j in final_indices:
        print(f'{count:3}) {int(yyyy_XC[mask][j])}-{int(mm_XC[mask][j]):02}-{int(dd_XC[mask][j]):02}/'\
              f'{int(hh_XC[mask][j]):02}:{int(mn_XC[mask][j]):02} | d_eq = {d_eq[mask][j]:6.2f} | '
              f'Juno_MLAT = {MLAT_Juno_trim[mask][j]:6.2f} | Juno_MLT = {MLT_Juno_trim[mask][j]:6.2f} | '
              f'Juno_MShell = {MSHELL_Juno[mask][j]:6.2f}')
        count+=1
            
    print('\n')
```
